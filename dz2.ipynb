```python
import numpy as np
import keras
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
```

```python
(X_train,y_train_labels),(X_test,y_test_labels) = fashion_mnist.load_data()
X_train.shape
```

```python
for ind in range(10):
    plt.imshow(X_train[ind])
    plt.show()
```

```python
from keras.layers import Dense
from keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train = X_train.reshape((-1,28*28))
X_test = X_test.reshape((-1,28*28))
```

```python
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
from keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train_labels)
y_test = to_categorical(y_test_labels)
```

```python
import math
def models(num_of_layers):
    model = Sequential()
    model.add(Dense(28,input_shape = (784,),activation = 'relu'))
    for n in range(1,num_of_layers):
        model.add(Dense(math.comb(num_of_layers, n) + 28,activation = 'relu'))
    model.add(Dense(10,activation = 'sigmoid'))
    return model
```

```python
def model2lay(nn):
    model = Sequential()
    model.add(Dense(28,input_shape = (784,),activation = 'relu'))
    model.add(Dense(nn,activation = 'relu'))
    model.add(Dense(10,activation = 'sigmoid'))
    return model
```

```python
%%time
ac = []
loss = []
for c in  range(1,100):
    print(c)
    m = model2lay(c)
    m.compile(optimizer = 'Adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
    m.fit(X_train,y_train,batch_size = 2048,epochs = 5,verbose = 1)
    ac.append(m.evaluate(X_test,y_test,verbose = 0)[1])
    loss.append(m.evaluate(X_test,y_test,verbose = 0)[0])
```

```python
plt.figure(figsize = (15,10))
plt.subplot(121)
plt.plot(range(1,100),loss)
plt.xlabel('neurons')
plt.ylabel('loss')
plt.subplot(122)
plt.plot(range(1,100),ac)
plt.xlabel('neurons')
plt.ylabel('accuracy')
```

```python
ind_max = np.argmax(np.array(ac))
ind_max + 1,ac[ind_max],loss[ind_max]
print(f'  : {ind_max + 1},  ac: {ac[ind_max]},  : {loss[ind_max]}')
```

```python
from tensorflow.keras.optimizers import *
```

```python
def opt_test(model,epochs,X_train,y_train,X_test,y_test,metrics = [SGD(),RMSprop(),Adam()]):
    fig,(ax1,ax2) = plt.subplots(1,2)
    ax1.figure.set_size_inches(10, 10)
    ax2.figure.set_size_inches(10, 10)
    colors = ['g', 'b', 'r']
    optimizers = metrics
    model.save_weights('weights.h5')
    models = {}
    for i,opt in enumerate(optimizers):
        model.load_weights('weights.h5')
        model.compile(loss = 'categorical_crossentropy',optimizer = opt,metrics = ['accuracy'])
        
        h0_tr = model.evaluate(X_train,y_train,verbose = 0)
        h0_te = model.evaluate(X_test,y_test,verbose = 0)
        
        hist = model.fit(X_train,y_train,epochs = epochs,batch_size = 512,verbose = 0,validation_data = (X_test,y_test))
        models.setdefault(opt.get_config()['name'],model)
        
        ax1.plot(np.arange(11),[h0_tr[0]] + hist.history['loss'],linestyle = '-',c=colors[i],label = opt.get_config()['name'] + ' train')
        ax1.plot(np.arange(11),[h0_te[0]] + hist.history['val_loss'],linestyle = '--',c=colors[i],label = opt.get_config()['name'] +' val')
        ax2.plot(np.arange(11),[h0_tr[1]] + hist.history['accuracy'],linestyle = '-',c=colors[i],label = opt.get_config()['name'] + ' train')
        ax2.plot(np.arange(11),[h0_te[1]] + hist.history['val_accuracy'],linestyle = '--',c=colors[i],label = opt.get_config()['name'] + ' test')
    ax1.set_xlabel('epochs')
    ax1.set_ylabel('loss')
    ax1.legend()
    ax2.set_xlabel('epochs')
    ax2.set_ylabel('accuracy')
    ax2.legend()
    return models
```

```python
%%time
model = models(2)
m2 = opt_test(model,10,X_train,y_train,X_test,y_test,metrics = [SGD(),RMSprop(),Adam()])
```

```python
%%time
model = models(3)
m3 = opt_test(model,10,X_train,y_train,X_test,y_test,metrics = [SGD(),RMSprop(),Adam()])
```

```python

def conf_mat(true,pred,cl):
    true_cl = np.ones(true.shape[0])
    pred_cl = np.ones(true.shape[0])
    for i in range(true.shape[0]):
        true_cl[i] = 1 if np.argmax(true[i]) == cl else 0
        pred_cl[i] = 1 if np.argmax(pred[i]) == cl else 0
    TP = np.logical_and(pred_cl == 1, true_cl == 1).sum()
    TN = np.logical_and(pred_cl == 0, true_cl == 0).sum()
    FP = np.logical_and(pred_cl == 1, true_cl == 0).sum()
    FN = np.logical_and(pred_cl == 0, true_cl == 1).sum()
    mat = [[TN,FN],
           [FP,TP]]
    
    return pred_cl,true_cl,np.array(mat)
```

```python

recall = []
prec = []
for i in range(y_test.shape[1]):
    pred_cl,true_cl,mat = conf_mat(y_test,pred,i)
    recall.append(mat[1,1]/(mat[1,1] + mat[0,1]))
    prec.append(mat[1,1]/(mat[1,1] + mat[1,0]))

    
prec = np.array(prec)
recall = np.array(recall)

recal_avg = np.mean(recall)
prec_avg = np.mean(prec)
print(f'recall_avg = {recal_avg}, precision_avg = {prec_avg} accuracy = {ac[ind]}')
```
